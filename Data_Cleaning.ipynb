```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\ADITYA GUPTA\Downloads\Data\messy_customer_dataset.csv")
df
```

```python
df["signup_date"] = pd.to_datetime(df["signup_date"], format='mixed', dayfirst=True).apply(lambda x: x if pd.notna(x) else pd.to_datetime("2024-01-01"))
df
```

```python
df['customer_id'] = df['customer_id'].astype(float)
df
```

```python
df['customer_id'] = df['customer_id'].interpolate()
df
```

```python
df['email'] = df['email'].fillna('provided@gmail.com')
df
```

```python
df['customer_id'] = df['customer_id'].astype(int)
df
```

```python
df['age'] = df['age'].fillna(df['age'].median())
df
```
